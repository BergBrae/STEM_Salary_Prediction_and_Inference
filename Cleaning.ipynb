```python
df = pd.read_csv('Levels_Fyi_Salary_Data.csv')
df.loc['Masters_Degree' : 'Race_Hispanic']
df.drop(['dmaid', 'otherdetails', 'rowNumber', 'cityid'], axis = 1, inplace = True)

df.drop(
    df.columns[
               np.where(df.columns.str.find('Masters_Degree') == 0)[0][0] :
               np.where(df.columns.str.find('Race_Hispanic') == 0)[0][0] + 1
    ], axis = 1, inplace = True
)


df['timestamp'] = pd.to_datetime(df['timestamp'])
df['state'] = df['location'].apply(lambda s: s.split(' ')[-1])
df['state'] = df['state'].apply(lambda s: s if len(s) ==2 else 'non US')
df
```

```python
fig = px.histogram(df, 'totalyearlycompensation')
x = df['totalyearlycompensation'].quantile(0.95)
print(x)
fig.show()
```
